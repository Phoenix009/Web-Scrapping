```python
from bs4 import BeautifulSoup
import requests
```

```python
source = requests.get("https://github.com/Phoenix009?tab=repositories").text
```

```python
soup = BeautifulSoup(source, "lxml")
```

```python
print(soup.prettify())
```

```python
repository = soup.find("div", class_ = "application-main")
repository = repository.find("main")
repository = repository.find("div", class_ = "col-lg-9 col-md-8 col-12 float-md-left pl-md-2")
repository = repository.find("div", class_ = "position-relative")
repository_list = repository.find("div", id = "user-repositories-list")
print(repository_list.prettify())
```

```python
repository_list = repository_list.find_all("a")
repository_list
```

```python
repo_name = []
repo_link = []
for repo in repository_list:
    try:
        if repo["class"]:
            pass
    except:
        repo_name.append(repo.text)
        repo_link.append("https://github.com"+repo["href"])
```

```python
for i in repo_name:
    print(i)
```

```python
for i in repo_link: 
    print(i)
```

```python
import csv
csv_file = open("repositories.csv", "w")
csv_writer = csv.writer(csv_file)
```

```python
csv_writer.writerow(["RepoName", "RepoLink"])
for i in range(len(repo_name)):
    csv_writer.writerow([repo_name[i], repo_link[i]])
csv_file.close()
```
